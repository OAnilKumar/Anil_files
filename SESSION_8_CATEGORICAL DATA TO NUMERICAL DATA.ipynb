# NP.WHERE

- IT IS APPLICABLE ONLY FOR BINARY

- NP.WHERE WILL TAKE THREE ARGUMENTS

    - CONDITION
 
    - TRUE VALUE
 
    - FALSE VALUE

$SENARIO 

- FOR EXAMPLE CASE STATUS VALUE EQUAL TO CERTIFIED THEN REPLACE WITH ZERO

- OTHER WISE REPLACE WITH 1

```python
visa_df=pd.read_csv("Visadataset - Visadataset.csv")
con=visa_df['case_status']=='Certified'
visa_df['case_status']=np.where(con,0,1)
visa_df
```

# ONE HEART ENCODER

- ONE HOT MEANS IF ONE IS 'ON' ANOTHER IS 'OFF'

- ON REPRESENTS WITH 1, AND OFF REPRESENTS WITH 0

- FOR EXAMPLE CASE STATUS HAS TWO UNIQUE CASES

    - DENIED
 
    - CERTIFIED
 
- IF WE APPLY ONE HOT ENCODER ON CASE STATUS COLUMN IT WILL CREATE TWO EXTRA COLUMNS

    - CASE_STATUS_DENIED
 
    - CASE_STATUS_CERTIFIED

|case_status | case_status_denied | case_status_certified|
|-----------|---|---|
|Denied|1|0|
|Certified|0|1|

**ADVANTAGES**

- IN MACHINE LEARNING WE HAVE ONE IMPORTANT PROPERTY IS **INDEPENDENT OF VARIBALES**

- ML MODELS EXPECTS VARIABLES ARE **INDEPENDENT TO EACH OTHER**

- ONE HOT ENCODER CREATING NEW COLUMNS , THE NEW COLUMNS ARE **INDEPENDENT TO EACHOTHER**

- INDEPENDENT VARIABLES MEANS BOTH VARIABLES HAVE **90 DEGREE PHASE SHIFT**

- 90 DEGREE PHASE SHIFT ALSO CALLED AS **PERPENDICULAR TO EACHOTHER**

- PERPENDICULAR TO EACHOTHER MEANS **ORTHOGONAL TO EACHOTHER**

**DISADVANTAGE**

- ONE HOT ENCODER CREATES NEW COLUMNS EQUAL TO NUMBER OF UNIQUE LABELS

- IF NUMBER OF UNIQUE LABELS ARE INCREASE ,NUMBER OF UNIQUE LABELS ALSO INCREASES

- THE MORE COLUMNS LEADS TO MORE DATA

- THEN DATA IS MORE THE COMPLEXITY WILL INCREASE

- COMPLEXITY INCREASES MORE COMPUTATION POWER REQUIRED

- ALSO TIME INCREASES

- THE ENTIRE NAME IS **CURSE OF DIMENTIONALITY**
