    - LOG TRANSFORMATION

    - EXPONENTIONAL TRANSFORMATION

# STEP-4

- **LOG TRANSFORMATION**

- LOG TRANSFORMATION MEANS PERFORMAING LOGARITHM OPERATIONS ON ORIGINAL DATA

- IT IS AN APPROACH TO CONVERT INTO NORMAL DISTRIBUTION

- LOG MEANS NATURAL LOGARITHM BASE = e


```python
log_data=np.log(exp_data)
log_data
```

```python
plt.hist(log_data,bins=20,label='Logarithm')
plt.show()
```

```python
plt.figure(figsize=(12,3))
plt.subplot(1,2,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=20,label='Logarithm')
plt.legend()
plt.show()
```

# POWER TRANSFORMATION

- BOX COX

- YE-JHONSON

- IT IS UNDER SKLEARN PREPROCESSING 

```python
from sklearn.preprocessing import PowerTransformer
power=PowerTransformer(method='yeo-johnson')
exp_data1=exp_data.reshape(-1,1)
data_trans=power.fit_transform(exp_data1)
plt.hist(data_trans)
```

```python
exp_data
```

```python
from sklearn.preprocessing import PowerTransformer
power=PowerTransformer(method='box-cox')
exp_data1=exp_data.reshape(-1,1)
box_cox_data_trans=power.fit_transform(exp_data1)
plt.hist(box_cox_data_trans)
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.subplot(1,2,1).hist(data_trans,bins=20,label='Yeo-Johnson')
plt.legend()
plt.subplot(1,2,2).hist(box_cox_data_trans,bins=20,label='Box-Cox')
plt.legend()
plt.show()
```
