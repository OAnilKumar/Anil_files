- SO IT IS IMPORTTANT TO KEEP ALL THE VALUES UNDER ONE SCALE

- WE HAVE TWO METHODS ARE THERE

    - STANDARDIZATION
 
    - NORMALIZATION
 
- STANDARDIZATION

# NORMALIZATION

```python
visa_df=pd.read_csv("Visadataset - Visadataset.csv")
col=visa_df['prevailing_wage']
min=visa_df['prevailing_wage'].min()
max=visa_df['prevailing_wage'].max()
numerator=visa_df['prevailing_wage']-min
denominator=max-min
visa_df['prevailing_wage_n']=numerator/denominator
visa_df
```

**package_name=MinMaxScalar**

```python
maxx=visa_df['prevailing_wage'].idxmax(),visa_df['prevailing_wage_n'].idxmax()
minn=visa_df['prevailing_wage'].idxmin(),visa_df['prevailing_wage_n'].idxmin()

maxx,minn
```
